## ğŸ“§ Email Sample Analysis

Loading and analyzing real email samples from the `emails_samples/` directory to demonstrate processing capabilities.

```python
# Load email samples from the email_samples folder
email_samples_dir = Path("../emails_samples")

def load_email_file(file_path):
    """Load and parse an .eml email file"""
    try:
        with open(file_path, 'rb') as f:
            msg = email.message_from_bytes(f.read(), policy=email.policy.default)
        
        # Extract basic info
        email_data = {
            'file_name': file_path.name,
            'sender': msg.get('From', 'Unknown'),
            'subject': msg.get('Subject', 'No Subject'),
            'date': msg.get('Date', 'Unknown'),
            'body': '',
            'html_body': '',
            'attachments': []
        }
        
        # Extract body content
        if msg.is_multipart():
            for part in msg.walk():
                content_type = part.get_content_type()
                if content_type == 'text/plain':
                    email_data['body'] = part.get_content()
                elif content_type == 'text/html':
                    email_data['html_body'] = part.get_content()
        else:
            email_data['body'] = msg.get_content()
            if msg.get_content_type() == 'text/html':
                email_data['html_body'] = email_data['body']
        
        # Use HTML body if plain text is empty
        if not email_data['body'] and email_data['html_body']:
            email_data['body'] = email_data['html_body']
        
        return email_data
        
    except Exception as e:
        print(f"âŒ Error loading {file_path}: {e}")
        return None

# Load sample emails
sample_emails = []
if email_samples_dir.exists():
    
    ### Select first 10 emails for demo
    # eml_files = list(email_samples_dir.glob('*.eml'))[:10]  # Load first 10 for demo
    
    ### Select 10 random emails for demo.
    ### Get ALL .eml files first
    all_eml_files = list(email_samples_dir.glob('*.eml'))
    
    # Select 10 RANDOM emails (or all if less than 10)
    if len(all_eml_files) >= 10:
        eml_files = random.sample(all_eml_files, 10)
        print(f"ï¿½ï¿½ Randomly selected 10 emails from {len(all_eml_files)} available")
    else:
        eml_files = all_eml_files
        print(f"ï¿½ï¿½ Using all {len(eml_files)} available emails")
    
    
    print(f"Loading {len(eml_files)} email samples...")
    
    for eml_file in eml_files:
        email_data = load_email_file(eml_file)
        if email_data:
            sample_emails.append(email_data)
    
    print(f"âœ… Loaded {len(sample_emails)} email samples successfully")
    
    # Show sample overview
    for i, email in enumerate(sample_emails[:3], 1):
        print(f"\nğŸ“§ Sample {i}:")
        print(f"  File: {email['file_name']}")
        print(f"  From: {email['sender'][:50]}...")
        print(f"  Subject: {email['subject'][:60]}...")
        print(f"  Body length: {len(email['body'])} characters")
else:
    print("âš ï¸ Email samples directory not found, using synthetic examples")
    # Create some synthetic examples for demo
    sample_emails = [
        {
            'file_name': 'demo_dbs_sales_scoop.eml',
            'sender': 'research@dbs.com',
            'subject': 'DBS SALES SCOOP - NVIDIA Upgrade',
            'body': 'NVIDIA Corporation (NVDA) - UPGRADED to BUY from HOLD. Target price raised to $520 from $480. Strong AI chip demand supports higher valuation. Download full analysis: https://research.dbs.com/nvidia_analysis.pdf',
            'html_body': '<html><body><h2>DBS SALES SCOOP</h2><p><strong>NVIDIA (NVDA) - BUY</strong></p><p>TP: $520</p><p><a href="https://research.dbs.com/nvidia_analysis.pdf">Full Report</a></p></body></html>',
            'attachments': []
        }
    ]

print(f"\nğŸ“Š Ready to process {len(sample_emails)} email samples")
```

## ğŸ¯ Dashboard Section 1: Signal Extraction Analysis

Extract and analyze trading signals (BUY/SELL/HOLD recommendations, target prices) from real financial emails.

```python
print("ğŸ¯ DASHBOARD SECTION 1: Signal Extraction Analysis")
print("=" * 60)
print("Analyzing trading signals from real financial emails...\n")

signal_results = []
total_signals_extracted = 0

for i, email in enumerate(sample_emails, 1):
    print(f"ğŸ“§ Processing Email {i}: {email['subject'][:50]}...")
    print(f"   From: {email['sender'][:40]}...")
    
    # Extract signals
    result = signal_extractor.extract_signals(email['body'])
    
    if result.has_signals:
        print(f"   âœ… SIGNALS FOUND: {len(result.signals)} signals (confidence: {result.extraction_confidence:.2f})")
        
        for j, signal in enumerate(result.signals, 1):
            signal_desc = f"{signal.signal_type.value.upper()}: {signal.action}"
            if signal.ticker:
                signal_desc += f" {signal.ticker}"
            if signal.value:
                signal_desc += f" {signal.value}"
            if signal.currency:
                signal_desc += f" {signal.currency}"
            
            print(f"      {j}. {signal_desc} (confidence: {signal.confidence:.2f})")
            print(f"         Context: {signal.context[:80]}...")
        
        total_signals_extracted += len(result.signals)
        
        # Store formatted result for dashboard analysis
        formatted_result = signal_extractor.format_signals_for_output(result)
        signal_results.append({
            'email': email,
            'signals': formatted_result
        })
    else:
        print(f"   âšª No trading signals found")
    
    print()  # Empty line
    
    if i >= 5:  # Limit to first 5 for dashboard demo
        break

print("=" * 60)
print(f"ğŸ“Š SIGNAL EXTRACTION DASHBOARD SUMMARY")
print(f"   ğŸ“§ Emails analyzed: {min(len(sample_emails), 5)}")
print(f"   ğŸ¯ Total signals extracted: {total_signals_extracted}")
print(f"   ğŸ“ˆ Emails with actionable signals: {len(signal_results)}")
print(f"   âœ… Signal detection rate: {len(signal_results)/min(len(sample_emails), 5)*100:.1f}%")
```

## ğŸ“Š Dashboard Section 2: Link Processing Intelligence

Analyze link extraction and research report harvesting capabilities from financial emails.

```python
print("ğŸ“Š DASHBOARD SECTION 2: Link Processing Intelligence")
print("=" * 60)
print("Analyzing link extraction and research report harvesting...\n")

async def dashboard_link_analysis():
    link_results = []
    total_links_found = 0
    total_research_reports = 0
    
    for i, email in enumerate(sample_emails[:3], 1):  # First 3 for dashboard demo
        print(f"ğŸ“§ Analyzing Email {i}: {email['subject'][:50]}...")
        
        # Use HTML body if available for better link extraction
        content_to_process = email.get('html_body', email['body']) or email['body']
        
        try:
            # Process links for dashboard analysis
            result = await link_processor.process_email_links(content_to_process)
            
            print(f"   ğŸ”— Links discovered: {result.total_links_found}")
            print(f"   ğŸ“Š Research reports identified: {len(result.research_reports)}")
            print(f"   ğŸŒ Portal links found: {len(result.portal_links)}")
            
            if result.research_reports:
                print(f"   ğŸ“¥ Research reports for processing:")
                for report in result.research_reports:
                    print(f"      â€¢ {report.url[:60]}...")
                    print(f"        Size: {report.file_size} bytes, Content: {len(report.text_content)} chars")
            
            if result.portal_links:
                print(f"   ğŸ”’ Portal access points identified:")
                for portal in result.portal_links[:3]:  # Show first 3
                    print(f"      â€¢ {portal.url[:60]}...")
            
            total_links_found += result.total_links_found
            total_research_reports += len(result.research_reports)
            
            link_results.append({
                'email': email,
                'link_result': result
            })
            
        except Exception as e:
            print(f"   âŒ Error in link processing: {e}")
        
        print()  # Empty line
    
    print("=" * 60)
    print(f"ğŸ“Š LINK PROCESSING DASHBOARD SUMMARY")
    print(f"   ğŸ“§ Emails analyzed: {len(link_results)}")
    print(f"   ğŸ”— Total links discovered: {total_links_found}")
    print(f"   ğŸ“Š Research reports identified: {total_research_reports}")
    print(f"   ğŸ’ Content enhancement capability: {'HIGH' if total_research_reports > 0 else 'MODERATE'}")
    
    return link_results

# Execute dashboard link analysis
link_dashboard_results = await dashboard_link_analysis()
```
